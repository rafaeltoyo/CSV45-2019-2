```python
from skimage import io, color, exposure
import glob
import numpy as np
from matplotlib import pyplot as plt
%pylab inline

d0 = []
d1 = []
d2 = []

for img in glob.glob("cidade/*.jpg"):    
    im=io.imread(img)
    print('Processando imagem ' + img)
    
    #conversão do espaço de cor - RGB para HSV
    im = color.rgb2hsv(im);  
   
    #descritor simples - soma de histograma
    aux1 = [x for x in im[:,:,0]]
    aux2 = [x/16 for x in im[:,:,1]]
    aux3 = [x/4 for x in im[:,:,2]]
    idx = np.add(aux1,np.add(aux2,aux3))
    d0.append(sum(idx))
    
    #média do V (valor - HSV)
    idx1 = np.mean(im[:,:,2])
    d1.append(idx1)    
    
    #média de um ajuste do V (valor - HSV)
    aux = [np.square(x*2) for x in im[:,:,2]]
    idx2 = np.mean(aux);
    d2.append(idx2)
  
#plots dos resultados
plt.figure()
plt.plot(d1[0:5], d0[0:5], 'ro')
plt.plot(d1[5:10], d0[5:10], 'b*')

plt.figure()
plt.plot(d2[0:5], d0[0:5], 'ro')
plt.plot(d2[5:10], d0[5:10], 'b*')
```
